```python
import numpy as np
import matplotlib.pyplot as plt

# Chức năng thu thập dữ liệu cho từng vật thể phóng
def thu_thap_du_lieu_vat_the(vat_the_phong):
    while True:
        try:
            so_lan_thu = int(input(f"Nhập số lần thử nghiệm cho {vat_the_phong}: "))
            so_lan_thanh_cong = int(input(f"Nhập số lần thành công cho {vat_the_phong}: "))
            if so_lan_thu < so_lan_thanh_cong or so_lan_thu < 1:
                raise ValueError("Số lần thử phải lớn hơn 0 và không nhỏ hơn số lần thành công!")

            khoang_cach = float(input(f"Nhập khoảng cách từ {vat_the_phong} đến mục tiêu (m): "))
            khoi_luong = float(input(f"Nhập khối lượng {vat_the_phong} (gam): "))
            return so_lan_thu, so_lan_thanh_cong, khoang_cach, khoi_luong

        except ValueError as e:
            print(f"Lỗi: {e}. Vui lòng nhập lại.\n")

# Chức năng tính toán độ chính xác
def tinh_toan_do_chinh_xac(so_lan_thu, so_lan_thanh_cong):
    if so_lan_thu == 0:
        return 0
    return (so_lan_thanh_cong / so_lan_thu) * 100

# Chức năng thu thập và so sánh dữ liệu giữa các vật thể phóng
def so_sanh_vat_the():
    while True:
        try:
            so_vat_the = int(input("Nhập số lượng vật thể phóng: "))
            if so_vat_the < 1:
                raise ValueError("Số lượng vật thể phải lớn hơn 0!")
            break
        except ValueError as e:
            print(f"Lỗi: {e}. Vui lòng nhập lại.\n")

    du_lieu_vat_the = []

    for i in range(so_vat_the):
        vat_the_phong = input(f"Nhập tên vật thể phóng thứ {i + 1}: ")
        so_lan_thu, so_lan_thanh_cong, khoang_cach, khoi_luong = thu_thap_du_lieu_vat_the(vat_the_phong)

        # Tính độ chính xác
        do_chinh_xac = tinh_toan_do_chinh_xac(so_lan_thu, so_lan_thanh_cong)
        du_lieu_vat_the.append({
            "vat_the_phong": vat_the_phong,
            "so_lan_thu": so_lan_thu,
            "so_lan_thanh_cong": so_lan_thanh_cong,
            "khoang_cach": khoang_cach,
            "khoi_luong": khoi_luong,
            "do_chinh_xac": do_chinh_xac
        })

    return du_lieu_vat_the

# Chức năng hiển thị biểu đồ so sánh độ chính xác giữa các vật thể
def hien_thi_bieu_do_so_sanh(du_lieu_vat_the):
    vat_the = [d["vat_the_phong"] for d in du_lieu_vat_the]
    do_chinh_xac = [d["do_chinh_xac"] for d in du_lieu_vat_the]

    plt.figure(figsize=(10, 6))
    plt.bar(vat_the, do_chinh_xac, color='blue', alpha=0.7)
    plt.xlabel("Vật thể phóng")
    plt.ylabel("Độ chính xác (%)")
    plt.title("So sánh độ chính xác giữa các vật thể phóng")
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# Chương trình chính
def main():
    du_lieu_vat_the = so_sanh_vat_the()

    # Hiển thị kết quả
    print("\n===== Kết quả so sánh độ chính xác giữa các vật thể phóng =====")
    for du_lieu in du_lieu_vat_the:
        print(f"Vật thể phóng: {du_lieu['vat_the_phong']}")
        print(f"Số lần thử: {du_lieu['so_lan_thu']}, Số lần thành công: {du_lieu['so_lan_thanh_cong']}")
        print(f"Khoảng cách: {du_lieu['khoang_cach']} mét, Khối lượng: {du_lieu['khoi_luong']} gam")
        print(f"Độ chính xác: {du_lieu['do_chinh_xac']:.2f}%\n")

    # Hiển thị biểu đồ so sánh độ chính xác
    hien_thi_bieu_do_so_sanh(du_lieu_vat_the)

if __name__ == "__main__":
    main()
```

```python
# Chức năng thu thập dữ liệu cho từng vật thể phóng
def thu_thap_du_lieu_vat_the(vat_the_phong, khoang_cach, so_lan_thu):
    while True:
        try:
            khoi_luong = float(input(f"Nhập khối lượng {vat_the_phong} (gam): "))
            return {
                "vat_the_phong": vat_the_phong,
                "khoang_cach": khoang_cach,
                "so_lan_thu": so_lan_thu,
                "khoi_luong": khoi_luong
            }
        except ValueError:
            print("Dữ liệu không hợp lệ! Vui lòng nhập lại.\n")

# Chức năng thu thập dữ liệu dự đoán cho các vật thể
def thu_thap_du_lieu_vat_the_du_doan(so_vat_the, khoang_cach, so_lan_thu):
    du_lieu_vat_the = []
    for i in range(so_vat_the):
        vat_the_phong = input(f"Nhập tên vật thể phóng thứ {i + 1}: ")
        du_lieu = thu_thap_du_lieu_vat_the(vat_the_phong, khoang_cach, so_lan_thu)
        du_lieu_vat_the.append(du_lieu)
    return du_lieu_vat_the

# Chức năng dự đoán độ chính xác (giả sử khối lượng nhẹ hơn có độ chính xác cao hơn)
def du_doan_do_chinh_xac(du_lieu_vat_the):
    khoi_luong_max = max(d["khoi_luong"] for d in du_lieu_vat_the)
    for du_lieu in du_lieu_vat_the:
        du_lieu["do_chinh_xac"] = 100 - (du_lieu["khoi_luong"] / khoi_luong_max) * 100

# Chức năng hiển thị biểu đồ dự đoán độ chính xác
def hien_thi_bieu_do_du_doan(du_lieu_vat_the):
    vat_the = [d["vat_the_phong"] for d in du_lieu_vat_the]
    do_chinh_xac = [d["do_chinh_xac"] for d in du_lieu_vat_the]

    plt.figure(figsize=(10, 6))
    plt.bar(vat_the, do_chinh_xac, color='green', alpha=0.7)
    plt.xlabel("Vật thể phóng")
    plt.ylabel("Dự đoán độ chính xác (%)")
    plt.title("Dự đoán độ chính xác cho các vật thể")
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# Chức năng chính để dự đoán độ chính xác cho các vật thể chưa thử nghiệm
def du_doan_vat_the():
    try:
        so_vat_the = int(input("Nhập số lượng vật thể muốn dự đoán: "))
        so_lan_thu = int(input("Nhập số lần thử dùng chung cho các vật thể: "))
        khoang_cach = float(input("Nhập khoảng cách chung cho các vật thể (m): "))

        du_lieu_vat_the = thu_thap_du_lieu_vat_the_du_doan(so_vat_the, khoang_cach, so_lan_thu)
        du_doan_do_chinh_xac(du_lieu_vat_the)

        # Hiển thị kết quả dự đoán
        print("\n===== Dự đoán độ chính xác cho các vật thể =====")
        for du_lieu in du_lieu_vat_the:
            print(f"Vật thể: {du_lieu['vat_the_phong']}, Khối lượng: {du_lieu['khoi_luong']} gam")
            print(f"Khoảng cách: {du_lieu['khoang_cach']} mét, Số lần thử: {du_lieu['so_lan_thu']}")
            print(f"Dự đoán độ chính xác: {du_lieu['do_chinh_xac']:.2f}%\n")

        # Hiển thị biểu đồ dự đoán
        hien_thi_bieu_do_du_doan(du_lieu_vat_the)

    except ValueError as e:
        print(f"Lỗi: {e}. Vui lòng nhập lại.\n")

# Chương trình chính
def main():
    lua_chon = input("Bạn có muốn dự đoán độ chính xác cho các vật thể chưa thử nghiệm? (có/không): ").strip().lower()
    if lua_chon == "có":
        du_doan_vat_the()
    else:
        print("Kết thúc chương trình.")

if __name__ == "__main__":
    main()
```
